```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
from pylab import rcParams
rcParams['figure.figsize'] = (12, 9)
```

```python
data=pd.read_excel('geo.xlsx')
```

```python
from sklearn.cluster import DBSCAN
```

Подбираем параметры DBscan

```python
#db=DBSCAN(eps=0.005, n_jobs=8, min_samples=1)
db=DBSCAN(eps=0.01, n_jobs=8, min_samples=1)
```

```python
%%time
clusters_all=db.fit_predict(data)
```

```python
plt.scatter(data['x'], data['y'], c=clusters_all, cmap='gray');
```

Теперь точки с меньшей плотностью объединим в один кластер, чтобы они отображались одним цветом.
Для этого посчитаем все точки каждого кластера и если их допустим <4000, будем присваивать им номер 999

```python
unique, counts=np.unique(clusters_all, return_counts=True)
```

```python
for x in range(len(clusters_all)):
    if np.count_nonzero(clusters_all==clusters_all[x])<3000:
        clusters_all[x]=999       
```

```python
dict(zip(unique,counts))
```
