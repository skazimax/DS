```python
from pymystem3 import Mystem

m = Mystem()
def lemmatize(text, mystem=m):
    try:
        return "".join(m.lemmatize(text)).strip()  
    except:
        return " "

data_tx_lem = data_tx.text.apply(lemmatize)
```

Посмотрим, про какие санкции чаще всего пишут

```python
from pymystem3 import Mystem
def find_sanc(x):
    x=' '.join(re.findall(r'\w+(?=\s+санкция)',x))
    m=Mystem()
    res=m.analyze(x)
    names = []
    for i in res: 
        if 'analysis' in i:
            if i['analysis']:
                if 'A' in i['analysis'][0]['gr']:
                    name = (i['text'])
                    if not name  in names:
                        names.append(name)
    return names
```

```python
data_tx['sanctions']=data_tx.lem_text.apply(lambda x: ''.join(find_sanc(x)))
```

```python
data_join['ClosingPrice_bin']=(data_join.ClosingPrice >= data_join.ClosingPrice.shift(1))*1
```

```python
X_data_train=data_join[['lem_text']][data_join.date<'01-01-2016']
X_data_test=data_join[['lem_text']][data_join.date>='01-01-2016']
y_data_train=data_join[['ClosingPrice_bin']][data_join.date<'01-01-2016']
y_data_test=data_join[['ClosingPrice_bin']][data_join.date>='01-01-2016']
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import TruncatedSVD, LatentDirichletAllocation
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.metrics import *
```

```python
Classifier=[LogisticRegression(),SGDClassifier(max_iter=1000, tol=0.001),RandomForestClassifier(),MultinomialNB()]
```

```python
for x in Classifier:
    clf = Pipeline([
        ('vect', CountVectorizer(ngram_range=(3,4), analyzer='char')),
        ('tfidf', TfidfTransformer()),
        #('reducer', LatentDirichletAllocation(n_components=50)),
        #('reducer', TruncatedSVD(n_components=50)),
        ('clf', x),
    ])
    clf.fit(X_data_train.lem_text, y_data_train.ClosingPrice_bin)
    preds=clf.predict(X_data_test.lem_text)
    print(*re.findall(r'\w+(?=\()',str(x)),':')
    print("Precision: {0:6.2f}".format(precision_score(y_data_test, preds)))
    print("Recall: {0:6.2f}".format(recall_score(y_data_test, preds)))
    print("F1-measure: {0:6.2f}".format(f1_score(y_data_test, preds)))
    print("Accuracy: {0:6.2f}\n".format(accuracy_score(y_data_test, preds)))
```

**Попробуем обучиться только на существительных**. <br/>Как видно, вышло столько же Accuracy - 0.51

```python
from pymystem3 import Mystem
m=Mystem()
```

```python
def get_subj(x):
    s=[]
    t=m.analyze(x)
    for i in range(len(t)):
        if 'analysis' in t[i]:
            try:
                if 'gr' in t[i]['analysis'][0]:
                    if 'S' in t[i]['analysis'][0]['gr']:
                        s.append(t[i]['text'])
            except:
                pass
    return " ".join(s)
```

```python
data_join['subj_only']=data_join.lem_text.apply(lambda x: get_subj(x)) 
```

```python
X_data_train=data_join[['subj_only']][data_join.date<'01-01-2016']
X_data_test=data_join[['subj_only']][data_join.date>='01-01-2016']
y_data_train=data_join[['ClosingPrice_bin']][data_join.date<'01-01-2016']
y_data_test=data_join[['ClosingPrice_bin']][data_join.date>='01-01-2016']
```

```python
Classifier=[LogisticRegression(),SGDClassifier(max_iter=1000, tol=0.001),RandomForestClassifier(),MultinomialNB()]
```

```python
for x in Classifier:
    clf = Pipeline([
        ('vect', CountVectorizer(ngram_range=(3,4), analyzer='char')),
        ('tfidf', TfidfTransformer()),
        #('reducer', TruncatedSVD(n_components=500))
        ('reducer', LatentDirichletAllocation(n_components=50)),
        ('clf', x),
    ])
    clf.fit(X_data_train.subj_only, y_data_train.ClosingPrice_bin)
    preds=clf.predict(X_data_test.subj_only)
    print(*re.findall(r'\w+(?=\()',str(x)),':')
    print("Precision: {0:6.2f}".format(precision_score(y_data_test, preds)))
    print("Recall: {0:6.2f}".format(recall_score(y_data_test, preds)))
    print("F1-measure: {0:6.2f}".format(f1_score(y_data_test, preds)))
    print("Accuracy: {0:6.2f}\n".format(accuracy_score(y_data_test, preds)))
```
