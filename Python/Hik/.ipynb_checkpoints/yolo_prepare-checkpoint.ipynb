```python
import os, glob
from distutils.dir_util import copy_tree
import random
```

```python
output_dir = '/mnt/Data/HIK/Out/'
model_dir = '/mnt/Data/HIK/Model/'
model_data = '/mnt/Data/HIK/Model/data/obj/'
```

```python
def copy_dirs(src, dst):
    src_fld = os.listdir(src)
    for fld in src_fld:
        copy_tree(src + fld, model_dir)
```

```python
copy_dirs(output_dir, model_dir)
```

```python
for x in glob.iglob(model_data + '*.jpg'):
    print (x)
```

```python
files = glob.glob(model_data + '*.jpg')
```

```python
random.shuffle(files)
```

```python
test = files[:int(len(files)*0.1)]
train = files[int(len(files)*0.1):]
```

```python
file_train = open('/mnt/Data/HIK/Model/data/train.txt', 'w')
file_test = open('/mnt/Data/HIK/Model/data/test.txt', 'w')
```

```python
for f in test:
    file_test.write(f + '\n')
for f in train:
    file_train.write(f + '\n')
```
