```python
import pandas as pd
from matplotlib import pyplot
%matplotlib inline
from plotly.offline import init_notebook_mode, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
data_dir='C:/Users/mrasskazov/Desktop/DS/Data/NLP'
data_tx=pd.read_csv(data_dir+'/texts.csv')
data_pr=pd.read_csv(data_dir+'/gazprom_prices.csv', sep=';')
```

```python
data_pr['ClosingPrice']=data_pr.ClosingPrice.str.replace(',','.').astype('float64')
```

```python
data_pr['Date']=pd.to_datetime(data_pr['Date'])
data_tx['date']=pd.to_datetime(data_tx['date'])
```

```python
data_pr_gr=data_pr[['Date','ClosingPrice']].groupby(data_pr['Date'].dt.strftime('%Y/%m')).mean().reset_index()
```

```python
data_tx_gr=data_tx[['date','len']].groupby(data_tx['date'].dt.strftime('%Y/%m')).mean().reset_index()
```

```python
data_tx_gr.head()
```

```python
data_tx['len']=data_tx.text.apply(lambda x: len(x))
```

```python
trace0=go.Scatter(x=data_tx_gr['date'], y=data_tx_gr['len'], name='Text_Len')
trace1=go.Scatter(x=data_pr_gr['Date'], y=data_pr_gr['ClosingPrice'], name='Gas_Price')
data=[trace1,trace0]
layout={'title':'USD_RUB\BRENT Price'}
fig=go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```
