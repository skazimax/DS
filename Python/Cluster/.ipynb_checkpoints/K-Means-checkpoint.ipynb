```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sb
import numpy as np
from sklearn.datasets import make_moons
from sklearn.datasets import make_blobs
```

```python
X,y=make_moons(n_samples=200, noise=0.05, random_state=0)
X1,y1=make_blobs(n_samples=200, random_state=0, centers=2, cluster_std=1.4)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='autumn', s=60)
```

```python
plt.scatter(X1[:,0], X1[:,1], c=y, cmap='autumn', s=60);
```

# K-Means

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

```python
algos=[KMeans(n_clusters=2),AgglomerativeClustering(n_clusters=2), DBSCAN(eps=0.2)]
```

```python
for algo in algos:
    pred3=algo.fit_predict(X)
    plt.figure()
    plt.scatter(X[:,0], X[:,1], c=pred3, cmap='autumn', s=60)
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans_pred=kmeans.fit_predict(X1)
```

```python
plt.scatter(X1[:,0], X1[:,1], c=kmeans_pred, cmap='magma', s=60)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100)
```

```python
kmeans2=KMeans(n_clusters=2)
kmeans_pred2=kmeans2.fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=kmeans_pred2, cmap='magma', s=60)
plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],s=100);
```
