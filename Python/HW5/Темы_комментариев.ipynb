Лемматизируем слова

```python
import pymorphy2
from nltk.tokenize import word_tokenize
```

```python
class MorphProvider:
    def __init__(self):
        self.cache = {}
        self.morph = pymorphy2.MorphAnalyzer()
    
    def __call__(self, w):
        w = w.lower()
        cached = self.cache.get(w)
        if cached:
            return cached
        try:
            morphed = self.morph.parse(w)[0].normal_form
            self.cache[w] = morphed
            return morphed
        except:
            return None
    
    def morph_string(self, s):
        words = word_tokenize(s)
        return " ".join([self.__call__(w) for w in words])
```

```python
morph = MorphProvider()
```

```python
from tqdm import tqdm
```

```python
data["normalized"] = [morph.morph_string(t) for t in tqdm(data.comment)]
```

Считаем самые частые

```python
words = [w for w in " ".join(data.normalized).split() if len(w)>3]
```
