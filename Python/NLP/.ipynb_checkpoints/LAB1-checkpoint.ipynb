```python
data_tx['sanctions']=data_tx.lem_text.apply(lambda x: ''.join(find_sanc(x)))
```

```python
from pymystem3 import Mystem
def find_sanc(x):
    x=' '.join(re.findall(r'\w+(?=\s+санкция)',x))
    m=Mystem()
    res=m.analyze(x)
    names = []
    for i in res: 
        if 'analysis' in i:
            if i['analysis']:
                if 'A' in i['analysis'][0]['gr']:
                    name = (i['text'])
                    if not name  in names:
                        names.append(name)
    return names
```

```python
data_join['ClosingPrice_bin']=(data_join.ClosingPrice >= data_join.ClosingPrice.shift(1))*1
```

```python
X_data_train=data_join[['lem_text']][data_join.date<'01-01-2016']
X_data_test=data_join[['lem_text']][data_join.date>='01-01-2016']
y_data_train=data_join[['ClosingPrice_bin']][data_join.date<'01-01-2016']
y_data_test=data_join[['ClosingPrice_bin']][data_join.date>='01-01-2016']
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import TruncatedSVD
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.metrics import *
```

```python
Classifier=[LogisticRegression,SGDClassifier,RandomForestClassifier,MultinomialNB]
```

```python
for x in Classifier:
    clf = Pipeline([
        ('vect', CountVectorizer(ngram_range=(2,2))),
        ('tfidf', TfidfTransformer()),
        #('reducer', TruncatedSVD(n_components=500)),
        ('clf', x()),
    ])
    clf.fit(X_data_train.lem_text, y_data_train.ClosingPrice_bin)
    preds=clf.predict(X_data_test.lem_text)
    print(*re.findall(r'\w+(?=\'>)',str(x)),':')
    print("Precision: {0:6.2f}".format(precision_score(y_data_test, preds)))
    print("Recall: {0:6.2f}".format(recall_score(y_data_test, preds)))
    print("F1-measure: {0:6.2f}".format(f1_score(y_data_test, preds)))
    print("Accuracy: {0:6.2f}\n".format(accuracy_score(y_data_test, preds)))
```

```python
from __future__ import print_function

from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.layers import Embedding
from keras.layers import Conv1D, GlobalMaxPooling1D
from keras.datasets import imdb
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
```

```python
TEXT_LENGTH = 1800
VOCABULARY_SIZE = 250000
EMBEDDING_DIM = 100
DIMS = 250
MAX_FEATURES = 5000
batch_size = 32

nb_filter = 250
filter_length = 3
hidden_dims = 250
nb_epoch = 100
```

```python
tokenizer = Tokenizer(num_words=MAX_FEATURES)
tokenizer.fit_on_texts(X_data_train.lem_text)
sequences = tokenizer.texts_to_sequences(X_data_train.lem_text)
X_train = tokenizer.sequences_to_matrix(sequences, mode='count')
sequences = tokenizer.texts_to_sequences(X_data_test.lem_text)
X_test = tokenizer.sequences_to_matrix(sequences, mode='count')
```

```python
model = Sequential()
model.add(Dense(128, input_shape=(MAX_FEATURES,), activation = 'relu'))
model.add(Dropout(0.1))
model.add(Dense(1, activation = 'softmax'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_data_train.ClosingPrice_bin, epochs=nb_epoch, batch_size=batch_size,  validation_split=0.1)
```

```python
y_pred=model.predict(X_test)
```

```python
print("Precision: {0:6.2f}".format(precision_score(y_data_test.ClosingPrice_bin, y_pred)))
print("Recall: {0:6.2f}".format(recall_score(y_data_test.ClosingPrice_bin, y_pred)))
print("F1-measure: {0:6.2f}".format(f1_score(y_data_test.ClosingPrice_bin, y_pred)))
print("Accuracy: {0:6.2f}\n".format(accuracy_score(y_data_test.ClosingPrice_bin, y_pred)))
```
